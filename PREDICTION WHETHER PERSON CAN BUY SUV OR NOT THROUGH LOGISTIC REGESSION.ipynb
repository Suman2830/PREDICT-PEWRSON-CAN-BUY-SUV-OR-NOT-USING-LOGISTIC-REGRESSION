# PREDICTING THE PERSON CAN BUY SUV OR NOT USING LOGISTIC REGRESSION

```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing dataset

social_network=pd.read_csv("SocialNetworkAds.csv")
```

```python
social_network
```

```python
# Extracing the independent variable

X=social_network.iloc[:,[2,3]].values
```

```python
X
```

```python
# Extracting the dependent variable

Y=social_network.iloc[:,4].values
```

```python
Y
```

```python
# Visualising the datset by drawing a correlation map

sns.heatmap(social_network.corr())
```

```python
# Splitting the data set into Training and Testing dataset

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
# Fitting Logistic regression to training dataset

from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,Y_train)
```

```python
# Pedicting the test set results

Y_pred=classifier.predict(X_test)
Y_pred

```

```python
# Visualising the training set results

from matplotlib.colors import ListedColormap
X_set,Y_set=X_train,Y_train
X1,X2=np.meshgrid(np.arange(start=X_set[:,0].min()-1,stop=X_set[:,0].max() +1,step=0.01),
                  np.arange(start=X_set[:,1].min()-1,stop=X_set[:,1].max() +1,step=0.01))
plt.contour(X1,X2,classifier.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),
           alpha=0.75,cmap=ListedColormap(("red","green")))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i,j in enumerate (np.unique(Y_set)):
    plt.scatter(X_set[Y_set==j,0],X_set[Y_set==j,1],c=ListedColormap(("red","green"))(i),label=j)
    
plt.title("Logistic Regression(Training Set)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```

```python
# Confusion Matrix evaluation

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
cm
```
